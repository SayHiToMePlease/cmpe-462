```python
import matplotlib.pyplot as plt
import numpy as np

def plot_framework_concept():
    """
    Generates a schematic figure illustrating the Outlier Detection Framework
    based on SVM Slack Variables.
    """
    plt.figure(figsize=(8, 6))
    
    # 1. Create a dummy decision boundary and margins
    x = np.linspace(-1, 4, 100)
    w, b = 1, -1.5
    hyperplane = -x + 1.5   # w.x + b = 0
    margin_pos = -x + 2.5   # w.x + b = 1
    margin_neg = -x + 0.5   # w.x + b = -1
    
    # Plot lines
    plt.plot(x, hyperplane, 'k-', linewidth=2, label='Hyperplane (Decision Boundary)')
    plt.plot(x, margin_pos, 'k--', linewidth=1, label='Margin (+1)')
    plt.plot(x, margin_neg, 'k--', linewidth=1) # Negative margin

    # 2. Plot "Normal" Points (Safe inside the class)
    plt.scatter([2.5, 3.0, 3.2], [1.0, 0.5, 2.0], color='blue', s=100, label='Normal Points (xi=0)')
    
    # 3. Plot "Support Vectors" (On the margin)
    plt.scatter([1.5, 2.0], [1.0, 0.5], s=150, facecolors='none', edgecolors='k', linewidth=2, label='Support Vectors')
    plt.scatter([1.5, 2.0], [1.0, 0.5], color='blue', s=100) # Fill

    # 4. Plot the "Outlier" (Violating the margin)
    outlier_x, outlier_y = 0.5, 1.0 # Deep inside the wrong side
    plt.scatter([outlier_x], [outlier_y], color='red', marker='X', s=200, label='Outlier (High xi)')
    
    # Draw the Slack Variable (distance line)
    # Distance from margin_pos to outlier
    plt.arrow(outlier_x, outlier_y, 0.7, 0.7, head_width=0.1, color='red', length_includes_head=True)
    plt.text(0.8, 1.4, r'$\xi_i > 0$', fontsize=16, color='red', fontweight='bold')

    # Formatting
    plt.xlim(-0.5, 4)
    plt.ylim(-0.5, 4)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Proposed Framework: Outliers as Margin Violators')
    plt.legend(loc='upper right')
    plt.grid(True, linestyle=':', alpha=0.6)
    
    plt.tight_layout()
    plt.show()

# Run it to get your figure
plot_framework_concept()
```
